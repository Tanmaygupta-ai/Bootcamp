```python
# WRITE A CODE FOR PASS OR FAIL, PASS BE 40 OR ABOVE RUN THIS CODE ON A LIST OF STUDENT OF 5 STUDENTS
students = ["Aman", "Abhi", "Karan", "Vinay", "Rahul"]
marks = [90, 65, 18, 85, 35]
for i in range(len(marks) - 1):
  if marks[i] > 40:
    print(students[i] + " - " +  "PASS")
  else:
    print(students[i] + " - " + "FAIL")
```

```python
# CODE TO CHECK WHETHER NUMBER IS POSITIVE, NEGATIVE OR ZERO IN A GIVEN LIST OF NUMBER
numbers = [10, -5, 0, 20, -15]
for num in numbers:
  if num > 0:
    print(num, "is positive")
  elif num < 0:
    print(num, "is negative")
  else:
    print(num, "is zero")
```

```python
# STUDENT NAME AS INPUT

# BASED ON THEIR MARKS YOU HAVE TO CONSIDER A PRINT STATEMENT FOR "PERFORMENCE" MARKS IS MORE THAN 75 THEN PERFORMANCE IS STRONG MARKS IS EQUAL TO OR GREATER THAN 40, PERFORMANCE SHOULD PRINT NEED TO IMPORVEMENT LESS 20 THEN PRINT FAIL ,REPEAT CLASS

# ATENDANCE 75- WARNING LOW ATTENDANCE STUDY_HOUR LESS THAN 4 PRINT SHOULD COME WARNING NEEDS MORE STUDY TIME


students = [
    {
        "name": "Aman",
        "study_hours": 6,
        "attendance": 85,
        "marks": 90
    },
    {
        "name": "Abhi",
        "study_hours": 3,
        "attendance": 70,
        "marks": 65
    },
    {
        "name": "Karan",
        "study_hours": 2,
        "attendance": 60,
        "marks": 18
    },
    {
        "name": "Vinay",
        "study_hours": 5,
        "attendance": 90,
        "marks": 85
    },
    {
        "name": "Rahul",
        "study_hours": 1,
        "attendance": 50,
        "marks": 35
    }
]

search = input("Enter the name of the student: ")


for student in students:
  if student["name"] == search:
    marks = student["marks"]

    print("Marks:", student["marks"])
    if marks > 75:
        print("Performance is Strong")

    elif marks >= 40:
        print("Performance needs improvement")

    else:
        print("Fail - Repeat Class")


    if student["attendance"] < 75:
        print("Warning: Low Attendance")

    if student["study_hours"] < 4:
        print("Warning: Need more study time")
```

```python
transactions = [
    {"transaction_id": "TX101", "amount": 6200, "method": "Credit Card"},
    {"transaction_id": "TX102", "amount": 5500, "method": "Gift Card"},
    {"transaction_id": "TX103", "amount": 1200, "method": "Gift Card"},
    {"transaction_id": "TX104", "amount": 8000, "method": "Gift Card"},
    {"transaction_id": "TX105", "amount": 450, "method": "PayPal"}
]

flagged_transactions = []

for transaction in transactions:
    if transaction["amount"] > 5000 and transaction["method"] == "Gift Card":
        flagged_transactions.append(transaction["transaction_id"])

print(flagged_transactions)
```
